    -	Tập dữ liệu có thể thiếu một số biến quan trọng khác có thể ảnh hưởng đến biến mục tiêu như mùa, vị trí địa lý, hay các yếu tố khác.
2.   Kỹ thuật Support Vector Machine
-   Ưu điểm:
    -   Hiệu suất cao: SVM có khả năng xử lý hiệu quả các tập dữ liệu lớn và phức tạp. Thuật toán có thể xây dựng các siêu phẳng tốt để phân loại các lớp dữ liệu.
    -   Hiệu quả trong không gian chiều cao: SVM có khả năng xử lý dữ liệu có số chiều cao mà không gặp vấn đề về việc tăng đáng kể chi phí tính toán.
    -   Độ tin cậy cao: SVM xây dựng các siêu phẳng tối ưu để phân loại dữ liệu và đảm bảo giới hạn sai số.
    -   Tính linh hoạt: SVM hỗ trợ sử dụng các hàm kernel để phân loại dữ liệu phi tuyến.
-   Nhược điểm:
    -   Độ phức tạp tính toán: Trong một số trường hợp, việc huấn luyện mô hình SVM có thể tốn nhiều thời gian và tài nguyên tính toán.
    -   Nhạy cảm với nhiễu: SVM có thể bị ảnh hưởng bởi nhiễu trong dữ liệu và dễ dẫn đến overfitting nếu không được điều chỉnh đúng.
    -   Khó khăn trong việc chọn siêu tham số: Việc tinh chỉnh các siêu tham số của SVM như hàm kernel, độ ràng buộc và thông số điều chỉnh đòn bẩy là một thách thức và yêu cầu kiến thức và kinh nghiệm.
3.	 Kỹ thuật RandomForest 
-	Ưu điểm:
    -	Mô hình có độ chính xác cao: Random Forest được coi là một trong những mô hình phân loại và dự đoán tốt nhất trong machine learning vì nó có khả năng giảm overfitting và cung cấp độ chính xác cao.
    -	Khả năng xử lý tập dữ liệu lớn: Random Forest có thể xử lý tập dữ liệu lớn, có số lượng đặc trưng lớn và giải quyết các vấn đề về phụ thuộc tuyến tính.	
    -	Khả năng xử lý các đặc trưng bị khuyết dữ liệu: Random Forest có thể xử lý các đặc trưng bị khuyết dữ liệu mà không cần loại bỏ chúng.
    -	Khả năng xác định đặc trưng quan trọng: Random Forest có thể đánh giá độ quan trọng của các đặc trưng và xác định đặc trưng quan trọng trong việc phân loại và dự đoán.
    -	Khả năng xử lý nhiễu: Random Forest có khả năng xử lý nhiễu và giảm thiểu ảnh hưởng của dữ liệu nhiễu.
-   Nhược điểm:
    -	Đòi hỏi thời gian và tài nguyên tính toán: Random Forest có thể yêu cầu nhiều thời gian và tài nguyên tính toán để xây dựng và đào tạo mô hình.
    -	Không dễ hiểu: Random Forest không dễ hiểu và giải thích cho các nhà phân tích dữ liệu do quá trình xây dựng mô hình phức tạp.
    -	Không phù hợp cho các bài toán với dữ liệu dạng chuỗi hoặc dữ liệu thời gian: Random Forest không phù hợp cho các bài toán phân loại và dự đoán trên dữ liệu dạng chuỗi hoặc dữ liệu thời gian do không có tính chất liên tục và tuần tự. 
    -	Có thể bị overfitting nếu số lượng cây quá lớn hoặc các tham số không được điều chỉnh tốt.

        -	n_estimators: Là số lượng cây quyết định trong rừng. Nó càng lớn thì mô hình sẽ càng ổn định và độ chính xác cũng sẽ cải thiện.         -	criterion: Là hàm mục tiêu được sử dụng để đo lường chất lượng của một phép phân chia. Trong thuật toán Random Forest, hai hàm mục tiêu phổ biến là Gini Impurity và Entropy. Hàm mục tiêu được sử dụng sẽ ảnh hưởng đến cách cây được tạo ra và độ chính xác của mô hình
        -	max_depth: Là độ sâu tối đa của cây quyết định. Khi độ sâu của cây quá lớn, mô hình sẽ dễ bị overfitting. Nếu giá trị max_depth quá nhỏ, mô hình sẽ dễ bị underfitting. Do đó, giá trị max_depth cần được lựa chọn một cách cẩn thận để đạt được sự cân bằng giữa độ chính xác và độ phức tạp của cây.
        -	min_samples_split: Là số lượng mẫu tối thiểu được yêu cầu để phân chia một nút thành các nút con. Nếu một nút có số lượng mẫu ít hơn min_samples_split, nó sẽ không được phân chia và trở thành một lá của cây. Tăng giá trị min_samples_split sẽ giảm độ sâu của cây và giúp tránh tình trạng overfitting.

        -	df: Tập dữ liệu chứa các transaction được mô tả bởi các itemset.
        -	min_support=0.2: Ngưỡng support tối thiểu của các itemset để được coi là phổ biến. Trong trường hợp này, ngưỡng support được đặt là 0.2, có nghĩa là một itemset phổ biến nếu nó xuất hiện trong ít nhất 20% các transaction.
        -	use_colnames=True: Sử dụng tên cột để đặt tên cho các item trong frequent itemsets. 
    -	Kết quả trả về là một DataFrame chứa các frequent itemset phổ biến với support tương ứng của chúng, cho thấy có 90 frequent itemsets với support tối thiểu là 0.2.Cụ thể, các cột của DataFrame là support và itemsets. Trong đó, hỗ trợ của luật kết hợp là hỗ trợ của tập đầu vào và đầu ra, độ tin cậy là tỷ lệ giữa hỗ trợ của tập đầu vào và tập đầu ra, và lift là một chỉ số đánh giá mức độ phụ thuộc giữa tập đầu vào và đầu ra, so với tần suất xuất hiện của tập đầu ra độc lập.

-	Sự phân chia công việc là đồng đều và các thành viên vui vẻ nhận công việc và hoàn thành
2.	 Đánh giá của nhóm trưởng
-	Các thành viên tham gia làm việc tích cực, hoàn thành tốt nhiệm vụ được giao
-	Họp nhóm đúng giờ và luôn luôn nêu ý kiến cá nhân để đóng góp vào đồ án
-	Có sự hỗ trợ qua lại giữa các thành viên với nhau
3.	Ưu, nhược điểm và cách cải tiến quá trình làm việc nhóm
-	Ưu điểm:
    -	Tất cả các thành viên trong nhóm đều làm nên ai cũng hiểu rất rõ về đồ án cũng như tiến độ nhóm mình đang làm
    -	Mọi người có thể chia sẻ kiến thức, học hỏi qua lại lẫn nhau 
    -	Có thể sắp xếp thời gian linh hoạt của mỗi cá nhân để hoàn thành công việc được giao
-	Nhược điểm:
    -	Thiếu sự tương tác trực tiếp giữa các thành viên với nhau
    -	Làm việc online nên chất lượng hoàn thiện công việc của các thành viên có thể giảm sút
-	Cách cải tiến:
    -	Thiết lập các quy tắc và mục tiêu rõ ràng của nhóm
    -	Thường xuyên tổ chức các cuộc họp trực tuyến định kỳ để thảo luận về tiến độ, chia sẻ ý kiến và giải quyết vấn đề.
    -	Xây dựng môi trường làm việc thoải mái, tạo điều kiện để mọi thành viên cảm thấy thoải mái và tự tin trong quá trình làm việc nhóm
